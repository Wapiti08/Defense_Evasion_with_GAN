```python
from pathlib import Path
import shutil
```

```python
def agg_reports(src_paths, des_path):
    n = 0
    for path in src_paths:
        for file_path in path.rglob('report.json'):
            # print("Copy from {} to {}".format(file_path, des_path.joinpath("report"+str(n)+".json")))
            shutil.copy(file_path, des_path.joinpath("report"+str(n)+".json"))
            if n == 116:
                print(file_path)
            # record the order to reports
            n += 1
        
        
```

```python
dataset_raw_reports = "../dataset/raw_reports"
dataset_root = "../dataset"
```

```python
# copy the ransom
paths = ["crypto", "locker", "wann"]
des_path = Path(dataset_root).joinpath("mali")
src_paths = [Path(dataset_raw_reports).joinpath(path) for path in paths]

agg_reports(src_paths, des_path)
```

```python
# copy the normal 
paths = ['normal']
des_path = Path(dataset_root).joinpath("benign")
src_paths = [Path(dataset_raw_reports).joinpath(path) for path in paths]

agg_reports(src_paths, des_path)
```
