```python
import nltk
from nltk import FreqDist
from nltk.tokenize import word_tokenize, regexp_tokenize
import pandas as pd

nltk.download('punkt')

# define variables for wordbag
token_index = 1
wordbag = {}
```

```python
def tokens_gen(sen):
    return word_tokenize(sen)

def spe_tokens_gen(sen):
    if '\\' in sen:
        return sen.split("\\")
    elif ' ' not in sen:
        return sen
    else:    
        return word_tokenize(sen)
```

```python
def create_wordbag(sen_list):
    ''' create the wordbag models for url query parameters
    :param filename: the file with urls inside
    :param max: 
    '''
    tokens = []
    global wordbag
    global token_index
    for lines in sen_list:
        
        # for line in lines:
        #     if line is not None:
        #         # tokens.extend(tokens_gen(line))
        #         tokens.extend(spe_tokens_gen(line))

        line = ' '.join(lines)
        tokens.extend(word_tokenize(line))

    freddist = nltk.FreqDist(tokens)
    tokens = list(freddist.keys())
    for token in tokens:
        if token not in wordbag.keys():
            wordbag[token] = token_index
            token_index +=1
        else:
            continue
```

```python
strings_list = strs_df['strings']

create_wordbag(strings_list)
```

```python
def wordbag_vec(sens_list, labels_list):
    '''
    :param features_list: undecoded malicious payloads
    :param labels_list: label_encoded malicious types
    '''
    Y = []
    X = []
    X_lens = []
    for index, lines in enumerate(sens_list):
        # for line in lines:
        #     if isinstance(line, str):
        #         # tokens = tokens_gen(line)
        #         tokens = spe_tokens_gen(line)
        #         # define the vecs here
        #         vecs = []
        #         for token in tokens:
        #             if token in wordbag.keys():
        #                 vecs.append(wordbag[token])
        #             else:
        #                 vecs.append(-1)
                        
        #         X_lens.append(len(vecs))
        #         X.append(vecs)
        #         Y.append(labels_list[index])
    
        line = ' '.join(lines)
        tokens = tokens_gen(line)
        vecs = []
        for token in tokens:
            if token in wordbag.keys():
                vecs.append(wordbag[token])
            else:
                vecs.append(-1)
                
        X_lens.append(len(vecs))
        X.append(vecs)
        Y.append(labels_list[index])


    max_len = max(X_lens)
    for x in X:
        if len(x)!= max_len:
            x.extend((max_len-len(x))*[0])

    X = np.array(X)
    # return X, X_lens
    return X, np.array(Y)
```

```python
labels_list = strs_df['Label']
X, Y = wordbag_vec(strings_list, labels_list)
```

```python
print(X)
print(X.shape)
print(Y)
print(Y.shape)

from numpy import savetxt
savetxt("../dataset/matrix/X_str.csv", X)
savetxt("../dataset/matrix/Y_str.csv", Y)
```

```python
# build binary classification model

from keras import Sequential
from keras import layers 
from keras.callbacks import EarlyStopping
def create_model(input_dim):
    model = Sequential()
    model.add(layers.Dense(32, input_dim=input_dim, activation='relu'))
    model.add(layers.Dense(32, activation='relu'))
    model.add(layers.Dense(1, activation='sigmoid'))
    return model

# Create model and train
model = create_model(X.shape[1])
model.compile(loss='binary_crossentropy', 
               optimizer='adam', 
               metrics=['accuracy'])

# normalization
from sklearn.preprocessing import Normalizer

transformer = Normalizer().fit(X)
X = transformer.transform(X)
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=42)

history = model.fit(X_train, y_train, epochs=10, validation_split=0.2, shuffle=True, 
                batch_size=15, verbose=2, callbacks= [EarlyStopping('accuracy', patience=10)])
```

```python
# test the accuracy on test dataset
y_pre = model.predict_classes(X_test)
from sklearn.metrics import accuracy_score

accuracy_score(y_test, y_pre)
```
