```python
from pathlib import Path
import shutil

# define orignal dataset
normal_folder = "D:\\Malware Analysis\\Cuckoo Report\\normal"
locker_folder = "D:\\Malware Analysis\\Cuckoo Report\\locker"
crypto_folder = "D:\\Malware Analysis\\Cuckoo Report\\crypto"
wann_folder = "D:\\Malware Analysis\\Cuckoo Report\\wann"

target_folder = "E:\\open_source_projects\\Analysis_Ransome_with_GAN\\dataset\\raw_reports"

ori_folders  = [normal_folder, locker_folder, crypto_folder, wann_folder]
```

```python
# recursively find the report files and copy to current dataset folder
file_name = 'report.json'

for path in ori_folders:
    for file in Path(path).rglob("*report.json"):
        order = file.parent.parent.stem
        name = file.parent.parent.parent.stem
        dst = Path(target_folder).joinpath(name, order)
        if not dst.is_dir():
            dst.mkdir(parents=True, exist_ok=True)
        else:
            print("Copy from {} to {}".format(file, Path(dst).joinpath(file_name)))
            shutil.copyfile(file, Path(dst).joinpath(file_name).as_posix())
```

```python
import os
# calculate the efficient reports in total testing cases
def effi_rate(path):
    i,j = 0,0
    for _, dirnames, filenames in os.walk(path):
        if 'reports' in dirnames:
            j += 1
        else:
            pass
        if "report.json" in filenames:
            i +=1
        else:
            pass
    print("\nFor {} dataset:".format(Path(path).stem))
    print("There are available {} examples in {}, the efficient rate is {}".format(i,j, round(i/j,5)))
    return i/j, i
```

```python
# load the reports
for path in ori_folders:
    effi_rate(path)
```
