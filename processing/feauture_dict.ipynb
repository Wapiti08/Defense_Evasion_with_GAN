```python
import pickle
```

```python
with open('./dataset/features/raw_features.pickle', 'rb') as fr:
    fea_df = pickle.load(fr)
```

```python
# check the unique value for every column element
pro_fea_df = fea_df.drop(['ID', 'Label'], axis=1).compute()
```

```python
pro_fea_df.head()['dropped'][1]
```

```python
new_fea_dict = pro_fea_df.to_dict('list')
```

```python
# check the type of value inside key
print(new_fea_dict['regkey_opened'][0][0])
type(new_fea_dict['regkey_opened'][0][0])
```

```python
# create uni_fea_dict to save unique values in every key
uni_fea_dict = {}

for key in new_fea_dict.keys():
    uni_fea_dict[key] = []

for key in list(new_fea_dict.keys()):
  value_list = []
  for fea_list in new_fea_dict[key]:
      if len(fea_list) != 0:
        value_list.extend(fea_list)
      # print(list(set(uni_fea_dict[key])))
  uni_fea_dict[key].extend(list(set(value_list)))
```

```python
uni_fea_dict
```

```python
# create total feature set
import pandas as pd

feature_dict = {'Index': [], 
                'Key': [],
                'Value': []}
order = 0
for key, values in uni_fea_dict.items():
    for value in values:
        order += 1
        feature_dict['Key'].append(key)
        feature_dict['Value'].append(value) 
        feature_dict['Index'].append(order)
df = pd.DataFrame.from_dict(feature_dict, orient='columns').set_index('Index')
```

```python
df.to_csv('../dataset/features/features.csv')
```
